
# Stripplot using Seaborn in Python

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

	
sns.set(style = 'whitegrid')
tip = sns.load_dataset("tips")
	
sns.stripplot(x="day", y="total_bill", data=tip)

plt.show()
```

# Draw a single horizontal strip plot using Stripplot

```python

# use to set style of background of plot
sns.set(style = 'whitegrid')

# loading data-set
tips = sns.load_dataset("tips")
sns.stripplot(x=tips["total_bill"])
```

# Draw strip plot using jitter parameter
## jitter can be used to provide displacements along the horizontal axis, which is useful when there are large clusters of data points

```python

# use to set style of background of plot
sns.set(style = 'whitegrid')

# loading data-set
tips = sns.load_dataset("tips")

sns.stripplot(x="day", y="total_bill", data=tips, jitter=0.1)
```

# Draw outlines around the data points using linewidth

```python

# use to set style of background of plot
sns.set(style = 'whitegrid')

# loading data-set
tips = sns.load_dataset("tips")

sns.stripplot(y="total_bill", x="day", data=tips,
				linewidth=3)
```

```python

sns.stripplot(y="total_bill", x="day", data=tips,
                   linewidth=2,edgecolor='green')
```

# Draw strip plot using hue parameter

```python

# use to set style of background of plot
sns.set(style = 'whitegrid')

# loading data-set
tips = sns.load_dataset("tips")

sns.stripplot(x="sex", y="total_bill", hue="day", data=tips)
```

# Draw each level of the hue variable at different locations on the major categorical axis

```python
sns.stripplot(x="day", y="total_bill", hue="smoker",
				data=tips, palette="Set2", dodge=True)
```

# Plotting large points and different aesthetics With marker and alpha parameter

```python


sns.stripplot(x="day", y="total_bill", hue="smoker",
				data=tips, palette="Set1", size=20,
				marker="s", alpha=0.2)

plt.show()
```
